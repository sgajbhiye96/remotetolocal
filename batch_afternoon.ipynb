```python
#add item
#remove Item
#add cart
#checkout
#exit
cart=[]
while True:
    print("\n====Shoping Cart Menu=====")
    print("1.Add Items")
    print("2.Remove Items")
    print("3.View Cart")
    print("4.Checkout")
    print("5.Exit")

    choice=input("enter your choice (1-5)")

    if choice=="1":
        Item=input("Enter Item Name")
        price=float(input("Enter item price"))
        cart.append((Item,price))
        print("Item added")
    elif choice=="2":
        Item=input("Enter Item Name")
        for i in cart:
            if i[0]==Item:
                cart.remove(Item)
            else:
                print("Item not Found in cart")
    elif choice=="3":
        if not cart:
            print("Cart is empty")
        else:
            for i,(Item,price) in enumerate(cart,1):
                print(f"{i}.{Item}----{price}")
    elif choice=="4":
        total=0
        for Item,price in cart:
            total=total+price
        
        print("checkout itmes")
        for Item,price in cart:
            print(f"{Item}----{price}")
        print(f"Final Amount:{total}")
    elif choice=="5":
        print("Thank you for choosing Our product and Services")
    break
        
```

```python
def smart_div(func):
    def inner(x,y):
        if x<y:
            x,y=y,x
        return func(x,y)
    return inner
            
```

```python
@smart_div
def div(x,y):
    return x/y
```

```python
#Recursion

def fib(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
```

```python
fib(5)
```

```python
def factorial(n):
    if n<=1:
        return 1
    else:
        return n*factorial(n-1)

factorial(5)
```

```python


def fibonacci_recursive(n):
    """Calculates the nth Fibonacci number using a simple recursive approach."""
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Example usage
print(fibonacci_recursive(10)) # Returns 55. Note: This assumes F(0)=0 and F(1)=1.
```

```python
import numpy as np


x=[10,20,30,40,50,150]

np.quantile(x,0.25)
np.quantile(x,0.75)

def outlier(x):
    out=[]
    Q1=np.quantile(x,0.25)
    Q3=np.quantile(x,0.75)

    IQR=Q3-Q1
    upper=Q3+(1.5*IQR)
    lower=Q1-(1.5*IQR)

    for i in x:
        if i>upper or i<lower:
            out.append(i)
    return out
            
    
```

```python
x=[10,20,30,40,50,150]
y=[2.6,3,4.9,5,5,6,6,7.9,8,8.2,67]
outlier(x)
outlier(y)
```

```python
import numpy as np
from scipy import stats

# Example: One-sample t-test
# Null Hypothesis (H0): The mean weight of children is 32kg.
# Alternative Hypothesis (H1): The mean weight of children is not 32kg.

# Sample data
data = np.array([30, 31, 33, 35, 29, 32, 34, 30, 31, 33])
expected_mean = 32

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, expected_mean)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Conclusion based on p-value (e.g., significance level alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence that the mean weight is not 32kg.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the mean weight is different from 32kg.")
```

```python
from scipy import stats
import numpy as np

# Sample data (e.g., scores before and after an intervention)
before_scores = np.array([75, 80, 82, 78, 85, 88, 79, 81, 83, 86])
after_scores = np.array([80, 85, 85, 80, 87, 90, 82, 84, 86, 89])

# Perform the paired samples t-test
t_statistic, p_value = stats.ttest_rel(after_scores, before_scores, alternative='two-sided') 
# 'two-sided' for H1: means are not equal
# 'less' for H1: mean of 'after' is less than 'before'
# 'greater' for H1: mean of 'after' is greater than 'before'

print(f"Paired Samples t-test:")
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```

```python
from scipy.stats import f_oneway
import pandas as pd

# Example data
data = {
    'Group_A': [89, 89, 88, 78, 79],
    'Group_B': [93, 92, 94, 89, 88],
    'Group_C': [89, 88, 89, 93, 90]
}

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(data['Group_A'], data['Group_B'], data['Group_C'])

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a statistically significant difference between group means.")
else:
    print("Fail to reject the null hypothesis: No statistically significant difference between group means.")
```

```python
import scipy.stats as stats
import numpy as np

# Define observed and expected frequencies
observed_data = [8, 6, 10, 7, 8, 11, 9]  # Example observed frequencies
expected_data = [9, 8, 11, 8, 10, 7, 6]  # Example expected frequencies

# Perform the Chi-square Goodness of Fit Test
chi_square_statistic, p_value = stats.chisquare(observed_data, expected_data)

print(f"Chi-square test statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")

# Determine the critical value for comparison (e.g., at 5% significance level, df=len(observed_data)-1)
alpha = 0.05
degrees_of_freedom = len(observed_data) - 1
critical_value = stats.chi2.ppf(1 - alpha, df=degrees_of_freedom)

print(f"Critical value (alpha={alpha}, df={degrees_of_freedom}): {critical_value}")

# Make a decision based on the comparison
if chi_square_statistic > critical_value:
    print("Reject the null hypothesis: There is a significant difference between observed and expected frequencies.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between observed and expected frequencies.")

# Alternatively, using the p-value
if p_value <= alpha:
    print("Reject the null hypothesis (based on p-value): There is a significant difference.")
else:
    print("Fail to reject the null hypothesis (based on p-value): There is no significant difference.")
```
